```
# -*- coding: utf-8 -*-

from __future__ import print_function
import keras
from keras.datasets import cifar10
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Conv2D, MaxPooling2D
import os
from google.colab import files
import matplotlib.pyplot as plt

batch_size = 32
num_classes = 10
epochs = 100
#data_augmentation = False
num_predictions = 20
save_dir = os.path.join(os.getcwd(), 'saved_models')
model_name = 'keras_cifar10_trained_model.h5'

# The data, split between train and test sets:
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
print('x_train shape:', x_train.shape)
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')

# Convert class vectors to binary class matrices.
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)

# Dropout and Data augmentation
modelDropData = Sequential()
modelDropData.add(Conv2D(32, (3, 3), padding='same',input_shape=x_train.shape[1:]))
modelDropData.add(Activation('relu'))
modelDropData.add(Conv2D(32, (3, 3), padding='same'))
modelDropData.add(Activation('relu'))
modelDropData.add(MaxPooling2D(pool_size=(2, 2)))
modelDropData.add(Dropout(0.25))

modelDropData.add(Conv2D(64, (3, 3), padding='same'))
modelDropData.add(Activation('relu'))
modelDropData.add(Conv2D(64, (3, 3), padding='same'))
modelDropData.add(Activation('relu'))
modelDropData.add(MaxPooling2D(pool_size=(2, 2)))
modelDropData.add(Dropout(0.25))

modelDropData.add(Flatten())
modelDropData.add(Dense(512))
modelDropData.add(Activation('relu'))
modelDropData.add(Dropout(0.5))
modelDropData.add(Dense(num_classes))
modelDropData.add(Activation('softmax'))

# Dropout and no Data augmentation
modelDropNoData = Sequential()
modelDropNoData.add(Conv2D(32, (3, 3), padding='same',input_shape=x_train.shape[1:]))
modelDropNoData.add(Activation('relu'))
modelDropNoData.add(Conv2D(32, (3, 3), padding='same'))
modelDropNoData.add(Activation('relu'))
modelDropNoData.add(MaxPooling2D(pool_size=(2, 2)))
modelDropNoData.add(Dropout(0.25))

modelDropNoData.add(Conv2D(64, (3, 3), padding='same'))
modelDropNoData.add(Activation('relu'))
modelDropNoData.add(Conv2D(64, (3, 3), padding='same'))
modelDropNoData.add(Activation('relu'))
modelDropNoData.add(MaxPooling2D(pool_size=(2, 2)))
modelDropNoData.add(Dropout(0.25))

modelDropNoData.add(Flatten())
modelDropNoData.add(Dense(512))
modelDropNoData.add(Activation('relu'))
modelDropNoData.add(Dropout(0.5))
modelDropNoData.add(Dense(num_classes))
modelDropNoData.add(Activation('softmax'))

# No Dropout and Data augmentation
modelNoDropData = Sequential()
modelNoDropData.add(Conv2D(32, (3, 3), padding='same',input_shape=x_train.shape[1:]))
modelNoDropData.add(Activation('relu'))
modelNoDropData.add(Conv2D(32, (3, 3), padding='same'))
modelNoDropData.add(Activation('relu'))
modelNoDropData.add(MaxPooling2D(pool_size=(2, 2)))

modelNoDropData.add(Conv2D(64, (3, 3), padding='same'))
modelNoDropData.add(Activation('relu'))
modelNoDropData.add(Conv2D(64, (3, 3), padding='same'))
modelNoDropData.add(Activation('relu'))
modelNoDropData.add(MaxPooling2D(pool_size=(2, 2)))

modelNoDropData.add(Flatten())
modelNoDropData.add(Dense(512))
modelNoDropData.add(Activation('relu'))
modelNoDropData.add(Dense(num_classes))
modelNoDropData.add(Activation('softmax'))

# No Dropout and no Data augmentation
modelNoDropNoData = Sequential()
modelNoDropNoData.add(Conv2D(32, (3, 3), padding='same',input_shape=x_train.shape[1:]))
modelNoDropNoData.add(Activation('relu'))
modelNoDropNoData.add(Conv2D(32, (3, 3), padding='same'))
modelNoDropNoData.add(Activation('relu'))
modelNoDropNoData.add(MaxPooling2D(pool_size=(2, 2)))

modelNoDropNoData.add(Conv2D(64, (3, 3), padding='same'))
modelNoDropNoData.add(Activation('relu'))
modelNoDropNoData.add(Conv2D(64, (3, 3), padding='same'))
modelNoDropNoData.add(Activation('relu'))
modelNoDropNoData.add(MaxPooling2D(pool_size=(2, 2)))

modelNoDropNoData.add(Flatten())
modelNoDropNoData.add(Dense(512))
modelNoDropNoData.add(Activation('relu'))
modelNoDropNoData.add(Dense(num_classes))
modelNoDropNoData.add(Activation('softmax'))

        
# initiate RMSprop optimizer
opt = keras.optimizers.rmsprop(lr=0.0001, decay=1e-6)


# Let's train the model using RMSprop
modelDropData.compile(loss='categorical_crossentropy',
              optimizer=opt,
              metrics=['accuracy'])
modelDropNoData.compile(loss='categorical_crossentropy',
              optimizer=opt,
              metrics=['accuracy'])
modelNoDropData.compile(loss='categorical_crossentropy',
              optimizer=opt,
              metrics=['accuracy'])
modelNoDropNoData.compile(loss='categorical_crossentropy',
              optimizer=opt,
              metrics=['accuracy'])

x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255

historyDropData = 0
historyDropNoData = 0
historyNoDropData = 0
historyNoDropNoData = 0

#if not data_augmentation:
historyDropNoData = modelDropNoData.fit(x_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          validation_data=(x_test, y_test),
          shuffle=True)

historyNoDropNoData = modelNoDropNoData.fit(x_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          validation_data=(x_test, y_test),
          shuffle=True)
    
    
#else:
    # This will do preprocessing and realtime data augmentation:
datagen = ImageDataGenerator(
    featurewise_center=False,  # set input mean to 0 over the dataset
    samplewise_center=False,  # set each sample mean to 0
    featurewise_std_normalization=False,  # divide inputs by std of the dataset
    samplewise_std_normalization=False,  # divide each input by its std
    zca_whitening=False,  # apply ZCA whitening
    zca_epsilon=1e-06,  # epsilon for ZCA whitening
    rotation_range=0,  # randomly rotate images in the range (degrees, 0 to 180)
    # randomly shift images horizontally (fraction of total width)
    width_shift_range=0.1,
    # randomly shift images vertically (fraction of total height)
    height_shift_range=0.1,
    shear_range=0.,  # set range for random shear
    zoom_range=0.,  # set range for random zoom
    channel_shift_range=0.,  # set range for random channel shifts
    # set mode for filling points outside the input boundaries
    fill_mode='nearest',
    cval=0.,  # value used for fill_mode = "constant"
    horizontal_flip=True,  # randomly flip images
    vertical_flip=False,  # randomly flip images
    # set rescaling factor (applied before any other transformation)
    rescale=None,
    # set function that will be applied on each input
    preprocessing_function=None,
    # image data format, either "channels_first" or "channels_last"
    data_format=None,
    # fraction of images reserved for validation (strictly between 0 and 1)
    validation_split=0.0)

# Compute quantities required for feature-wise normalization
# (std, mean, and principal components if ZCA whitening is applied).
datagen.fit(x_train)
# Fit the model on the batches generated by datagen.flow().
historyDropData = modelDropData.fit_generator(datagen.flow(x_train, y_train,
                                 batch_size=batch_size),
                    steps_per_epoch=1563,
                    epochs=epochs,
                    validation_data=(x_test, y_test),
                    workers=4)

historyNoDropData= modelNoDropData.fit_generator(datagen.flow(x_train, y_train,
                                 batch_size=batch_size),
                    steps_per_epoch=1563,
                    epochs=epochs,
                    validation_data=(x_test, y_test),
                    workers=4)

f1 = plt.figure()
ax1 = f1.add_subplot(111)
ax1.set_title('Training Accuracy')
ax1.plot(historyDropData.history['acc']) 
ax1.plot(historyDropNoData.history['acc']) 
ax1.plot(historyNoDropData.history['acc']) 
ax1.plot(historyNoDropNoData.history['acc']) 
ax1.set_ylabel('Accuracy')
ax1.set_xlabel('Epoch')
ax1.legend(['Dropout and Data Augmentation', 'Dropout and no Data Augmentation',
           'No Dropout and Data Augmentation', 'No Dropout and no Data Augmentation'], loc='upper left')
f1.savefig("q1c_train.png")
files.download('q1c_train.png')

f2 = plt.figure()
ax2 = f2.add_subplot(111)
ax2.set_title('Test Accuracy')
ax2.plot(historyDropData.history['val_acc']) 
ax2.plot(historyDropNoData.history['val_acc']) 
ax2.plot(historyNoDropData.history['val_acc']) 
ax2.plot(historyNoDropNoData.history['val_acc']) 
ax2.set_ylabel('Accuracy')
ax2.set_xlabel('Epoch')
ax2.legend(['Dropout and Data Augmentation', 'Dropout and no Data Augmentation',
           'No Dropout and Data Augmentation', 'No Dropout and no Data Augmentation'], loc='upper left')
f2.savefig("q1c_test.png")

files.download('q1c_test.png')

# Save model and weights
# if not os.path.isdir(save_dir):
#     os.makedirs(save_dir)
# model_path = os.path.join(save_dir, model_name)
# model.save(model_path)
# print('Saved trained model at %s ' % model_path)

# # Score trained model.
# scores = model.evaluate(x_test, y_test, verbose=1)
# print('Test loss:', scores[0])
# print('Test accuracy:', scores[1])
```
